## Concatenate Function
- it is used for combining together Series, Dataframes and Panel Objects

#### pd.concat( ) 

    PARAMETER:    objs | axis | join | join_axes | ignore_index | keys | sort

```python
import pandas as pd
import numpy as np
```

```python
sr1 = pd.Series([0,1,2])
sr2 = pd.Series([3,4,5,6,7])
print(sr1, "\n\n", sr2)

pd.concat([sr1, sr2])   ## concatenatimg two "series"
```

```python
df_16 = pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[15,16,17,18]})
df_17 = pd.DataFrame({"ID":[5,6,7,8],"Name":["EE","FF","GG","HH"],"Class":[9,10,11,12]})
display(df_16, df_17,"-------------")

display(pd.concat([df_16, df_17])) ## concatenatimg two "DatFrames" 
pd.concat([df_16, df_17], axis = 1)   ## concatenatimg two "DatFrames"
```

```python
## Parameter : ignore_index

display(pd.concat([df_16, df_17], ignore_index = True))
```

```python
## Concate "Dataframes" with unequal elements

df_18 = pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[15,16,17,18]})
df_19 = pd.DataFrame({"ID":[7,8],"Name":["GG","HH"],"Class":[11,12]})
display(df_18,df_19,"-------------")

display(pd.concat([df_18, df_19])) 
pd.concat([df_18, df_19], axis = 1)
## NaN is displayed where vales in "df_19" dataframe is not available
```

```python
## Parameter : join

display(pd.concat([df_18, df_19], axis = 1),"Default")       ## Default is "Outer"
display(pd.concat([df_18, df_19], axis = 1, join = "outer"),"Outer") 

display(pd.concat([df_18, df_19], axis = 1, join = "inner"),"Inner")
## Displaying only common elements
```

```python
## Parameter : keys

display(pd.concat([df_18, df_19], keys = ["First df","second df"]))   
display(pd.concat([df_18, df_19],axis=1, keys = ["First df","second df"]))
```

```python
## Dataframes with different column names
df_21 = pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[15,16,17,18]})
df_22 = pd.DataFrame({"Marks":[57,78,69,80]})

display(df_21,df_22 )

pd.concat([df_21, df_22], sort=False)     ## Parameter: Sort
```

```python
df = pd.DataFrame(np.random.randn(8,4), index = ["Index"+str(i) for i in range(1,9)], columns = "A B C D".split())
display(df)     #Print Dataframe

List_df = [df[3:6],  df[:3]  ,df[6:]]    ## LIST of DataFrame
display(List_df)

pd.concat(List_df)  #concatenate a list into Dataframe
```
