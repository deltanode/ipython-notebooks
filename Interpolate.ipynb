# Interpolate & its various  techniques

**Syntax:** DataFrame.interpolate()

**Parameter:** method | axis | limit | inplace | limit_direction | limit_area

```python
import pandas as pd
```

```python
df = pd.read_csv("youtube_dataset\\glory.csv")
df
```

```python
## Linear (Default)

df.interpolate()                    
df.interpolate(method="linear")
```

```python
## Time

print(type(df["Date"][0]))     ## Verifying datatype of "Date" column 

df = pd.read_csv("youtube_dataset\\glory.csv", parse_dates=["Date"])    ## Changing datatype from "str to "Timestamp"
print(type(df["Date"][0]))

df.index = df["Date"]      ## Setting Index as Date
df.interpolate(method="time")
```

```python
## Index

df.index = [i for i in range(11,21)] ## Changing index
df.interpolate(method="index")
```

```python
## Nearest

df.interpolate(method="nearest")
```

```python
## Polynomial

# df.index = df["Date"]
df.interpolate(method="polynomial", order=2)
```

```python
## Spline

df.interpolate(method="spline", order=2)
```

```python
## Parameter: Limit & limit-direction

print(df)
# df.interpolate(limit=1, limit_directions="forward")
# df.interpolate(limit=1, limit_directions="backward")
df.interpolate(limit=1, limit_directions="both")
```
