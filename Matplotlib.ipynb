# Matplotlib

```python
## imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

## Line Plot

- #### plt.plot( )

```python
days = list(range(1,16))   # Generating random data List
d_temp = [round(random.uniform(30,42),2) for _ in range(15)]  # Generating random data List
m_temp = [round(random.uniform(25,42),2) for _ in range(15)]  

plt.plot(days,d_temp)
plt.title("Delhi Temprature")
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.legend

plt.show()
```

```python
## Setting the Axis Manually

days = list(range(1,16))
d_temp = [round(random.uniform(25,42),2) for _ in range(15)]  ## creating random data
m_temp = [round(random.uniform(25,42),2) for _ in range(15)]

plt.axis([0,18,0,50])
plt.plot(days,d_temp)
plt.title("Delhi Temprature")
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.legend

plt.show()
```

```python
## Setting the Axis Manually

days = list(range(1,16))
d_temp = [round(random.uniform(30,42),2) for _ in range(15)]  ## creating random data
m_temp = [round(random.uniform(20,35),2) for _ in range(15)]

# plt.plot(days,d_temp, color="g", marker="o", linestyle="--", linewidth="4", markersize="22")
plt.plot(days,d_temp, "ro--", linewidth="3", markersize="8", label="Delhi")
plt.plot(days,m_temp, "o:", linewidth="3", markersize="8", label="Mumbai")
plt.axis([0,18,10,50])
plt.title("Delhi & Mumbai Temprature", fontsize=14)
plt.xlabel("Days",fontsize=12)
plt.ylabel("Temprature",fontsize=12)
plt.legend(loc=1)
# plt.legend(["d_temp","m_temp"], loc=1)
plt.grid()

plt.show()
```

## Histogram

- #### plt.hist( )

```python
# Generating random data 

ml_student_age = np.random.randint(18,45,(100))  
py_student_age = np.random.randint(15,40, (100))

# display(ml_student_age,py_student_age)
```

```python
## Ploting Histogram & parameter : rwidth

# plt.hist((ml_student_age), label="Machine Learning")
plt.hist((ml_student_age),rwidth=0.8, label="Machine Learning")

plt.title("ML Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=2)

plt.show()
```

```python
## Parameter : Bins

bins = list(range(15,50,5))   ## Generating Bins

plt.hist((ml_student_age),bins,rwidth=0.8, label="Machine Learning")

plt.title("ML Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=2)

plt.show()
```

```python
## Parameter : histtype {"bar", "barsjtacked","steps","stepfilled"}

bins = list(range(15,50,5))   ## Generating Bins

plt.hist((ml_student_age),bins,rwidth=0.8, histtype="step", label="Machine Learning")

plt.title("ML Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=2)

plt.show()
```

```python
## Parameter : orienation {"vertical", "horizontal"}  & color

bins = list(range(15,50,5))   ## Generating Bins

plt.hist((ml_student_age),bins,rwidth=0.8, orientation="horizontal", color="g", label="Machine Learning")

plt.title("ML Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=4)

plt.show()
```

```python
## Changing Histogram Size

plt.figure(figsize= (10,6))           ## Figure Size

bins = list(range(15,50,5))   ## Generating Bins
plt.hist((ml_student_age,py_student_age),bins,rwidth=0.8, label=("Machine Learning","Python"))

plt.title("ML & Python Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=2)

plt.show()
```

```python
## Adding Style

from matplotlib import style
style.use("ggplot")


bins = list(range(15,50,5))   ## Generating Bins
plt.hist((ml_student_age,py_student_age),bins,rwidth=0.8, label=("Machine Learning","Python"))

plt.title("ML & Python Students Histogram", fontsize=14)
plt.xlabel("Age",fontsize=12)
plt.ylabel("No. of Students",fontsize=12)
plt.legend(loc=2)

plt.show()
```

## Bar Chart

- #### plt.bar( )

```python
## Data

classes = ["Python", "R", "AI", "ML", "DS"]

class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
## Plotting Bar Chart

plt.bar(classes,class1_students)
plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend("Class1 Students")
```

```python
## Horizontal Bar Chart

plt.barh(classes,class1_students)
plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend("Class1 Students")
```

```python
## Parameter : width | align | color | edgecolor | linewidth | alpha (Opacity) | linestyle |  visible | Label

plt.bar(classes,class1_students, width=0.6, align="edge", color="y", edgecolor = "orange", linewidth=5, 
        alpha=0.5, linestyle="--", visible=True, label = "Class1 Students")
plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend()
```

```python
## Ploting Bar Plot

w=0.2
classes_index = np.arange(len(classes))
# display(classes_index)

plt.bar(classes_index, class1_students, width=w, label="Class1 Students")
plt.bar(classes_index+w, class2_students, width=w, label="Class2 Students")
plt.bar(classes_index+w+w, class2_students, width=w, label="Class2 Students")

plt.xticks(classes_index+w,classes)  ## Setting X-Labels

plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

```python
## Ploting Bar Plot (Horizontal)

w=0.2
classes_index = np.arange(len(classes))
# display(classes_index)

plt.barh(classes_index, class1_students, w, label="Class1 Students")
plt.barh(classes_index+w, class2_students, w, label="Class2 Students")
plt.barh(classes_index+w+w, class2_students, w, label="Class2 Students")

plt.yticks(classes_index+w,classes)  ## Setting X-Labels

plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.legend()
plt.show()
```

# Scatter Plot

- #### plt.scatter()

```python
## Data

df_google_playstore = pd.read_csv("dataset\\playstore\\googleplaystore.csv")
df_google_playstore.shape
```

```python
df_google_playstore = pd.read_csv("dataset\\playstore\\googleplaystore.csv", nrows=1000)
df_google_playstore.shape
```

```python
df_google_playstore.head()
```

```python
x = df_google_playstore["Rating"]
y = df_google_playstore["Reviews"]
z= df_google_playstore["Installs"]
```

```python
## Plotting Scatter Plot

plt.scatter(x,y)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
## Parameter: |c (color) | Marker | s (Marker size) | alpha (opacity)| linewidth | edgecolor | verts (Similar to marker)

plt.figure(figsize=(8,5))

plt.scatter(x,y, c="r", marker="*", s=25, alpha=0.6, linewidth=4, edgecolor="g")
plt.title("Google Play Store Scatter Plot", fontsize=14)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Reviews", fontsize=12)
```

```python
## Plotting Scatter plot

plt.figure(figsize=(8,5))

plt.scatter(x,y,  marker="o", s=40, alpha=0.6)
plt.scatter(x,z, marker="*", s=40, alpha=0.6)
plt.title("Google Play Store Scatter Plot", fontsize=14)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Reviews & Installs", fontsize=12)
```
