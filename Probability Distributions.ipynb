# Distributions

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5,10,200)
# print(xs)
ks = np.arange(50)
# print(ks)
```

## Discrete PMFs

Binomial distribution and Poisson distribution

```python
pmf_binom = st.binom.pmf(ks, 50, 0.25)     ##@param events // chance of success
# print(pmf_binom)
plt.bar(ks,pmf_binom, label="Binomial Example", alpha=0.8)

pmf_poisson = st.poisson.pmf(ks, 30)        ##@param rates
# print(pmf_binom)
plt.bar(ks,pmf_poisson, label="Poisson Example", alpha=0.8)
plt.legend()


## What is the chance that u roll ten 1's on when you have 50 in total
print(st.binom.pmf(10, 50, 0.25))
## If we have characteristically 30 car crashes a night. what is  the chance that we got 50.
print(st.poisson.pmf(50, 30) )
```

## Continious Distributions

Uniform, normal, exponential,student-t,log normal, skew normal


```python
pdf_uniform = st.uniform.pdf(xs, -4, 10)
plt.plot(xs, pdf_uniform, label="Uniform(-4,6)")

pdf_normal = st.norm.pdf(xs, loc=5, scale=2)
plt.plot(xs, pdf_normal, label="Normal(5, 2)")

pdf_exponential = st.expon.pdf(xs, loc=-2, scale=2)
plt.plot(xs, pdf_exponential, label="Exponential(0.5)")

pdf_studentt = st.t.pdf(xs, 1)
plt.plot(xs, pdf_studentt, label="Student-t(1)")

pdf_lognorm = st.lognorm.pdf(xs, 1)
plt.plot(xs, pdf_lognorm, label="Lognorm(1)")

pdf_skewnorm = st.skewnorm.pdf(xs, -6)
plt.plot(xs, pdf_skewnorm, label="Skewnorm(5)")

plt.legend()
plt.ylabel("Prob")
plt.xlabel("x");
```
